# Hexbin plot

````python
from sklearn.datasets import load_iris
df = pd.DataFrame(load_iris().data, columns = load_iris().feature_names)
ax = df.plot.hexbin(x = 'petal length (cm)', y = 'sepal length (cm)', 
                    gridsize= 10,
                    sharex=False,
                    title = 'Count',
                    cmap='Greys')
````

# Bubble plot

````python
x = np.array([1,10,30,22,5,20])
y = np.array([4,17,4,33,20,15])
z = np.array([20,15,15,34,33,30])*10
t = np.array(["T1", "T2", "T1", "T2", "T2", "T2"])
label = [15,20,25,30]

fig = plt.scatter(x, y,
            s = z,
            c = "red",
            alpha = 0.5)
plt.xlabel("x")
plt.ylabel("y")

plt.title("Bubble chart", loc="left")

plt.legend(*fig.legend_elements(prop = "sizes", 
                                alpha = 0.5, 
                                num = 4, 
                                c = "red"),
                                loc="upper right", 
                                title="Sizes")


for i, txt in enumerate(t):
    plt.text(x[i]-.5, y[i]-.5, txt)

plt.show()
````
