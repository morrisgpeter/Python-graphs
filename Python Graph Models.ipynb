```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Hexbin

```python
from sklearn.datasets import load_iris
df = pd.DataFrame(load_iris().data, columns = load_iris().feature_names)
ax = df.plot.hexbin(x = 'petal length (cm)', y = 'sepal length (cm)', 
                    gridsize= 10,
                    sharex=False,
                    title = 'Count',
                    cmap='Greys')
```
